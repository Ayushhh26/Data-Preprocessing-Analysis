# USING GroupBy in Pandas


## Q.) Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in ascending order WRT ratings.

```python
pdf = df[df['Category']=='ART_AND_DESIGN']
pdf = pdf[pdf['Rating']>4.5]
pdf.sort_values( by='Rating', ascending = True)
```

## Q.) Which category has the maximum average rating

```python
df.groupby('Category').mean()['Rating'].sort_values(ascending =False)
```

## Q.)How many free apps are there in each category

```python
df1 = df[df['Type']=='Free']
df1.groupby('Category').count()['Type'].sort_values(ascending =False)
```
